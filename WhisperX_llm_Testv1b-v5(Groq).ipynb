**Run the Whisper model to generate text transcript**

````python
#print into dialog format
dialog =""

audio_prompt=f"""
The audio is the conversation between a telemarketer and customer.Carefully understand the audio. Identitfy the speakers relative to the audio speech.
Assign the correct speaker label to each speech.
In highly ambiguous cases, make your **best guess** based on the overall flow of the conversation. Be consistent with speaker assignment throughout the dialogue.

**No Strict Turn-Taking**: The dialogue may not follow a simple back-and-forth pattern. A single speaker may speak multiple consecutive sentences. Use meaning and flow to determine the correct speaker.
Then generate a transcript with speakers label. Output the transcription in a dialogue format with speaker label.
If the consersation is in other languages, translate to english transcript with speakers label.
"""

# Transcribe the audio
# Open the audio file

audio_model="distil-whisper-large-v3-en"
# audio_model="whisper-large-v3-en" # cannot use in for free

with open(audio_file, "rb") as file:
    # Create a transcription of the audio file
    transcription = client.audio.transcriptions.create(
      file=(audio_file, file.read()), # Required audio file
      model= audio_model, # Required model to use for transcription
      prompt="audio_prompt",  # Optional
      response_format="json",  # Optional
      language="en",  # Optional
      temperature=0.0  # Optional
    )
    # Print the transcription text
    print(transcription.text)

````

**Run LLM to do speaker label for the transcript**

````python

user_message = f"""
**Task: Speaker Labeling for a Two-Speaker Conversation with Ambiguity Handling**

You are given a transcript of a conversation between two speakers. Your task is to:

1. **Understand the Content**: Carefully analyze the sentences in the transcript to fully understand the context and flow of the conversation.

2. **Assign Speaker Labels**: Using your understanding of the dialogue, assign one of two speaker labels: "Speaker 1" and "Speaker 2". Ensure each sentence is labeled according to the correct speaker.

3. **Handle Ambiguity**: If you encounter sentences where the speaker is not immediately clear, use context clues such as:
   - **Tone and phrasing**: Who is likely to say a particular phrase based on tone (e.g., polite, formal vs. casual, informal).
   - **Response patterns**: How the other speaker responds to or acknowledges a previous statement.
   - **Content knowledge**: Information or expertise that only one speaker might possess (e.g., a telemarketer explaining a product).

   In highly ambiguous cases, make your **best guess** based on the overall flow of the conversation. Be consistent with speaker assignment throughout the dialogue.

4. **No Strict Turn-Taking**: The dialogue may not follow a simple back-and-forth pattern. A single speaker may speak multiple consecutive sentences. Use meaning and flow to determine the correct speaker.

5. **Output**: Provide the labeled transcript with correct speaker assignments.

---

#### Example Output:
```
Speaker 1: I’m having some trouble understanding the service offer you mentioned earlier.
Speaker 2: Sure, I can explain that. We’re offering a special discount for our premium package.
Speaker 1: Does it include faster support?
Speaker 2: Yes, you’ll have access to 24/7 customer support with priority response times.
Speaker 1: That makes sense. I think I’m interested in signing up.
Speaker 2: Great! Let me walk you through the process.
```

---

#### Key Considerations:
- **Contextual Understanding**: Use the meaning of each sentence and context clues to ensure accurate speaker identification.
- **Handling Ambiguity**: In cases of uncertainty, make a **consistent and logical decision** based on the flow and content of the conversation.
- **Non-Alternating Dialogue**: Be ready to label consecutive sentences by the same speaker if necessary.

---

<<<
Transcript: {transcription.text}
>>>

"""

dialog = run_groq(user_message)
print(dialog)
````
